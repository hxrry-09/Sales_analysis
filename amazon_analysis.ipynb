Let's start with these cleaning tasks.

```python
import gensim
from gensim.utils import simple_preprocess
from gensim.parsing.preprocessing import STOPWORDS
from nltk.stem import WordNetLemmatizer, SnowballStemmer
from nltk.stem.porter import *

stemmer = SnowballStemmer("english")
def lemmatize_stemming(text):
    return stemmer.stem(WordNetLemmatizer().lemmatize(text, pos='v'))

def preprocess(text):
    result = []
    for token in gensim.utils.simple_preprocess(text):
        if token not in gensim.parsing.preprocessing.STOPWORDS and len(token) > 3:
            result.append(lemmatize_stemming(token))
    return result

processed_reviews = df['review_content'].map(preprocess)

dictionary = gensim.corpora.Dictionary(processed_reviews)
dictionary.filter_extremes(no_below=15, no_above=0.5, keep_n=100000)

bow_corpus = [dictionary.doc2bow(doc) for doc in processed_reviews]

lda_model = gensim.models.LdaMulticore(bow_corpus, num_topics=10, id2word=dictionary, passes=2, workers=2)

for idx, topic in lda_model.print_topics(-1):
    print('Topic: {} \nWords: {}'.format(idx, topic))
```

We can try to make a machine learning model to predict the rating of a product based on the features.
Features I'm selecting are as following:
discounted_price, actual_price, discount_percentage, and review_length.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

features = ['discounted_price', 'actual_price', 'discount_percentage', 'review_length']
target = 'rating'
X_train, X_test, y_train, y_test = train_test_split(df[features], df[target], test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

mse, r2
```

The performance of the model is evaluated using two metrics:

Mean Squared Error (MSE): This is the average of the squares of the differences between the actual and predicted values. The lower the MSE, the better the model's performance. In this case, the MSE is approximately 0.079.

R-squared (R²): This is the proportion of the variance in the dependent variable that is predictable from the independent variables. It ranges from 0 to 1, with 1 indicating that the model perfectly predicts the target variable and 0 indicating that the model does not predict the target variable at all. In this case, the R² is approximately 0.039.

Given the low R² value, this model does not explain much of the variance in the ratings. This suggests that the features we included (discounted_price, actual_price, discount_percentage, and review_length) do not have strong predictive power for the product ratings. This aligns with our earlier exploratory analysis, which showed that these features had low correlation with the ratings.

Let us try to improve the performance of the model by adding more features. We can add the product category as a feature, since we saw earlier that the category appears to have some relationship with the ratings.

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

df['main_category'] = df['category'].apply(lambda x: x.split('|')[0])

features = ['discounted_price', 'actual_price', 'discount_percentage', 'review_length', 'main_category']
target = 'rating'

preprocessor = ColumnTransformer(
    transformers=[
        ('num', 'passthrough', ['discounted_price', 'actual_price', 'discount_percentage', 'review_length']),
        ('cat', OneHotEncoder(handle_unknown='ignore'), ['main_category'])
    ])

model = RandomForestRegressor(random_state=42)

pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', model)
])
X_train, X_test, y_train, y_test = train_test_split(df[features], df[target], test_size=0.2, random_state=42)

pipeline.fit(X_train, y_train)

y_pred = pipeline.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

mse, r2
```

The Mean Squared Error (MSE) of the model has decreased to approximately 0.065 from 0.079, and the R-squared (R²) value has increased to approximately 0.208 from 0.039.

This means that the model's performance has improved after including the main_category feature and switching to a Random Forest Regressor. The higher R² value indicates that the model now explains about 20.8% of the variance in the ratings, compared to only 3.9% before.

We can also try to predict if a product will have a high rating using the following steps:

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

median_rating = df['rating'].median()

df['high_rating'] = (df['rating'] > median_rating).astype(int)
features = ['discounted_price', 'actual_price', 'discount_percentage', 'review_length', 'main_category']
target = 'high_rating'
preprocessor = ColumnTransformer(
    transformers=[
        ('num', 'passthrough', ['discounted_price', 'actual_price', 'discount_percentage', 'review_length']),
        ('cat', OneHotEncoder(handle_unknown='ignore'), ['main_category'])
    ])

model = RandomForestClassifier(random_state=42)
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', model)
])
X_train, X_test, y_train, y_test = train_test_split(df[features], df[target], test_size=0.2, random_state=42)
pipeline.fit(X_train, y_train)
y_pred = pipeline.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

accuracy, precision, recall, f1
```

Accuracy is the proportion of correct predictions (both true positives and true negatives) among the total number of cases examined. In this context, it's the proportion of products for which the model correctly predicted whether they have a high rating or not.

Precision is the proportion of true positives (products correctly predicted as having a high rating) among all positives predicted by the model. It's a measure of the model's ability to correctly identify products with a high rating.

Recall (or sensitivity) is the proportion of true positives (products correctly predicted as having a high rating) among all actual positives. It's a measure of the model's ability to find all the products that have a high rating.

The F1 Score is the harmonic mean of precision and recall. It provides a single score that balances both the concerns of precision and recall in one number
